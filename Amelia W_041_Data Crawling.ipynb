```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.ruparupa.com/informastore/c/dekorasi/dekorasi-rumah/tanaman-artifisial.html?itm_source=mini-category-category-tanaman-artifisial&itm_campaign=third-level&itm_device=desktop')
```

```python
BeautifulSoup(page.content, 'html.parser')
```

```python
Soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#nyari satu kutipan saja
Soup.find('span', class_='product__name').text
```

```python
#nyari satu kutipan saja
Soup.find('div', class_='price__real').text
```

```python
##cari banyak kutipan
produk_html = Soup.find_all('span', class_='product__name')
```

```python
for each_produk_html in produk_html:
    print(each_produk_html.text)
```

```python
html_price = Soup.find_all('div', class_='price__real')
```

```python
for each_html_price in html_price:
    print(each_html_price.text)
```

```python
data_produk = []
for each_produk_html in produk_html:
    data_produk.append(each_produk_html.text)
```

```python
data_produk
```

```python
data_price = []
for each_html_price in html_price:
    data_price.append(each_html_price.text)
```

```python
data_price
```

```python
data = {
    'data_produk': data_produk,
    'data_price': data_price
}
```

```python
pd.DataFrame(data)
```
